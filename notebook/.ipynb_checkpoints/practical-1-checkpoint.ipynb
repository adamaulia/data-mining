Find two features where a certain group has a noticeable high mortality rate.

Distributions are pretty straightforward to interpret; the visualization between two features is simple and effective. However, it also limits displaying interaction between several features (e.g., if young males tended to have lower survival rates than females). For this, we can use `scatter plots`.

To stack multiple subsets in one plot and give them different colors, we must select them from the DataFrame based on some conditions. For example, we can do this for the `Survived` feature using the following piece of code:

Scatterplots work well when we have continuous features. However, the visualization can be misleading if at least one feature is nominal. In such cases, we should use barplots or boxplots. Next, we will show how to use boxplots to create informative visualizations.

Roughly speaking, a boxplot is a standardized way of displaying the distribution of data based on a five-number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”). Boxplots can tell us about the outliers in the data and what their values are. They can also tell us if the data is symmetrical, how tightly your data is grouped, and if and how the data is skewed.

```python
# let us try with Pclass and Survived that are two continuous features
ax=df.boxplot(column=['Pclass'], by='Survived', showbox=True, showmeans=True, grid=False)
ax=plt.suptitle('')
```

### Task 3. 
Set `Age` to $x$, and `Fare` to $y$. Do these features correlate?

The Pearson correlation coefficient measures the linear relationship between two continuous variables. Strictly speaking, Pearson’s correlation requires that each dataset be normally distributed. Correlation values vary between -1 and +1, with 0 implying no correlation. Correlations of -1 or +1 imply an exact linear relationship. Positive correlations imply that as $x$ increases, so does y. Negative correlations imply that as $x$ increases, $y$ decreases.

The p-value roughly indicates the probability of an uncorrelated system producing datasets with a Pearson correlation value at least as extreme as the one computed from these datasets. The p-values are not entirely reliable but are probably reasonable for large datasets.

<div>

</div>

```python
from scipy.stats import pearsonr
corr, p_value = pearsonr(df['Age'], df['Fare'])

print('The correlation is ' + str(corr))
print('The p-value is ' + str(p_value))
```

### Task 4.

Determine a likely ticket price to survive the Titanic.

The intuition dictates that we should search for outliers in the data concerning passengers who died after having paid higher price tickets. Later on, we can just compute the likely ticket price average of these outliers. The outliers can be visualized with a boxplot.



```python
df.boxplot(column=['Fare'],by='Survived', showbox=True, showmeans=True, grid=False)
ax=plt.suptitle('')
```

We can also determine the exact ticket price using the following piece of code. I have to the following method is not straightforward in my opinion. Therefore, saying that “the passenger was an outlier if you paid more than 50 and died” is perfectly acceptable.

The box will range from the 25th-percentile to 75th-percentile, and the whisker will range from the smallest value to the largest value between (25th-percentile - 1.5xIQR, 75th-percentile + 1.5xIQR), where IQR denotes the inter-quartile range. Of course, the 1.5 value is customizable.

<div>

</div>

```python
df_died = df[df['Survived'] == 0]
Q1 = df_died['Fare'].quantile(0.25)
Q3 = df_died['Fare'].quantile(0.75)

IQR = Q3 - Q1

Lower_Fence = Q1 - (1.5 * IQR)
Upper_Fence = Q3 + (1.5 * IQR)

l = min(df_died['Fare'][df_died['Fare']>Lower_Fence])
u = max(df_died['Fare'][df_died['Fare']<Upper_Fence])

print('Likely ticket price: ' + str(u)) # this is the exact value you are searching for
```
